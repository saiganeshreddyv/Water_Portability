```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import pandas as pd
df=pd.read_csv('/content/water_potability.csv')
df
```

```python
import numpy as np
x=np.mean(df['ph'])
print(x)
df['ph'].fillna(7, inplace = True)
df
```

```python
import numpy as np
import statistics as st
x=st.mode(df['Hardness'])
df["Hardness"].fillna(x, inplace = True)
```

```python
x=np.median(df['Turbidity'])
df["Turbidity"].fillna(x, inplace = True)
x=np.mean(df['Trihalomethanes'])
df["Trihalomethanes"].fillna(x, inplace = True)
x=np.median(df['Organic_carbon'])
df["Organic_carbon"].fillna(x, inplace = True)
x=np.median(df['Conductivity'])
df["Conductivity"].fillna(x, inplace = True)
x=np.mean(df['Sulfate'])
df["Sulfate"].fillna(320, inplace = True)
x=np.median(df['Chloramines'])
df["Chloramines"].fillna(x, inplace = True)
x=np.median(df['Solids'])
df["Solids"].fillna(x, inplace = True)
df
```

```python
f=df
f=f.drop('Potability',axis=1)
f=f.drop('Chloramines',axis=1)
f=f.drop('Sulfate',axis=1)
f=f.drop('Organic_carbon',axis=1)
f=f.drop('Trihalomethanes',axis=1)
f=f.drop('Conductivity',axis=1)
f=f.drop('Turbidity',axis=1)
f=f.drop('Hardness',axis=1)
f=f.drop('Solids',axis=1)
f
```

```python
x=df.drop('Potability',axis=1)
y=df['Potability']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42)
```
